### 作业1    某城市ncp情况

```python
import requests
import json

provinceName=input('输入您感兴趣的省份')

cityName=input('输入您感兴趣的城市')

url='https://wuliang.art/ncov/statistics/getCityHistoryList?cityName= %s &provinceName= %s '%(cityName,provinceName)

response = requests.get(url)          #爬虫·············（从网页里获得内容)
print(response)
```

```python
datas = response.text                 #把内容转化为字符串型·····(变成json文件)
#print(datas)

for row in datas["data"]:
    print(row['date'],end='\t')
    print(row['confirm'],end='\t')
    print(row['city'],end='\t')
    print(row['heal'],end='\t')
    print(row['dead'],end='\t')
    print(row['suspect'],end='\t')
    print(row['confirm_add'])
```

```python
pythondataobject = json.loads(datas)  #从json文件变成python文件···(字符串型变字典型)
#print(pythondataobject)
type(pythondataobject)
```

### 作业2 某城市每日新闻

```python
import requests
import json

cityName=input('输入您感兴趣的城市')

url='https://wuliang.art//ncov/news/getCityNewsList?cityName=%s'%(cityName)

response = requests.get(url)          #爬虫·············（从网页里获得内容)

datas = response.text                 #把内容转化为字符串型·····(变成json文件)

pythondataobject = json.loads(datas)  #从json文件变成python文件···(字符串型变字典型)

#print(pythondataobject['code'])

for row in pythondataobject['data']:
    b=pythondataobject['data']
    for row_2 in b['result']:
        c=b['result']
        for row_3 in c['govern']:
            print(row_3['ftime'],'\t')
            print(row_3['wx_name'],'\t') 
            print(row_3['rank_score'],'\t') 
            print(row_3['img'],'\t') 
            print(row_3['second_cls'],'\t') 
            print(row_3['title'],'\t') 
            print(row_3['ptime'],'\t') 
            print(row_3['url'],'\t') 
    for row_2_2 in b['tab']:
        print(row_2_2['name'],'\t') 
        print(row_2_2['key'],'\t') 
    print(b['city'],'\t')
    print(b['district'],'\t')
    print(b['query'])
        
```

### 作业3 口罩信息

```python
import requests
import json


url='https://wuliang.art//ncov/mask/gauzeMask'

response = requests.get(url)          #爬虫·············（从网页里获得内容)

datas = response.text                 #把内容转化为字符串型·····(变成json文件)

pythondataobject = json.loads(datas)  #从json文件变成python文件···(字符串型变字典型)


for row in pythondataobject['data']:
    print(row['total'],end='\t')
    print(row['province'],end='\t')
    print(row['isLoc'],end='\t')
    for row_2 in row['list']:
        print(row_2['orderUrl'],end='\t')
        print(row_2['total'],end='\t')
        print(row_2['city'],end='\t')
        print(row_2['isLoc'],end='\t')
        for row_3 in row_2['list']:
            print(row_3['district'],end='\t')
            print(row_3['url'])

```

### 作业4  辟谣专区（保存到csv）

```python
import requests
import json

for i in range(37):
    
    url='https://wuliang.art/ncov/rumor/getRumorList?page=%d '%(i)

response = requests.get(url)          #爬虫·············（从网页里获得内容)

datas = response.text                 #把内容转化为字符串型·····(变成json文件)

pythondataobject = json.loads(datas)  #从json文件变成python文件···(字符串型变字典型)


for row in pythondataobject['data']:
    info=''
    info += row['date']+','
    info += row['explain']+','
    info += row['arttype']+','
    info += row['author']+','
    info += row['section']+','
    info += row['abstract']+','
    info += row['title']+','
    info += str(row['type'])+','
    info += row['coversqual']+','
    info += row['result']+','
    info += row['cover']+','
    info += str(row['iscolled'])+','
    info += row['videourl']+','
    info += row['authordesc']+','
    info += row['coverrect']+','
    info += row['markstyle']+','
    info += str(row['id'])+','
    for row_2 in row['tag']:
        info += row_2
    info +='\n'
    with open('./result.csv','a') as f:
        f.write(info)
print(info)

  
```
