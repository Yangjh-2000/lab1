### 分词
0. 增加8个表的表头
1. 建立专业、证书、softskill表
2. 遍历这个三个表，并根据条件存入职位表dataframe中
3. 将dataframe导出新表格

```python
import pandas as pd

# 查看原职位表（产品）
df_poduct = pd.read_csv('产品.csv',engine='python')
df_poduct.head()
```

```python
# 新建的专业表
df_zy = pd.read_excel('V2_zy.xlsx')
df_zy.head()
```

```python
# 新建的证书表
df_book = pd.read_excel('V2_book.xlsx')
df_book.head()
```

```python
# 新建的能力表
df_ability = pd.read_excel('V2_ability.xlsx')
df_ability.head()
```

```python
# 新建的softskill表
df_softskill = pd.read_excel('V2_sofskill.xlsx')
df_softskill.head()
```

```python
# 测试
# 把专业表内容放入产品表

for i in range(10000):
    try:
        zy_total = ''
        for zy in df_zy['word']:
            if zy in df_poduct.loc[i,'职位描述']:
                zy_total += zy +','
                df_poduct.loc[i,'专业要求']=zy_total
    except:
        pass
    continue
```

```python
# 建立函数

def add(df_job):
    # 从第一行开始添加
    for i in range(32000):
        try:
            
            # 遍历专业表
            zy_total = ''
            for zy in df_zy['word']:
                # 与jd有相同词时添加
                if zy in df_job.loc[i,'职位描述']:
                    zy_total += zy +','
                    df_job.loc[i,'专业要求']=zy_total
                    
            # 遍历证书表        
            book_total = ''
            for book in df_book['word']:
                if book in df_job.loc[i,'职位描述']:
                    book_total += book +','
                    df_job.loc[i,'证书要求']=book_total
             
            # 遍历能力表
            ability_total = ''
            for ability in df_ability['word']:
                if ability in df_job.loc[i,'职位描述']:
                    ability_total += ability +','
                    df_job.loc[i,'技能要求（同职位关键词）']=ability_total
                    
            # 遍历softskill表
            softskill_total = ''
            for softskill in df_softskill['word']:
                if softskill in df_job.loc[i,'职位描述']:
                    softskill_total += softskill +','
                    df_job.loc[i,'softskill要求']=softskill_total
                    
                    
        except:
            pass
        continue
```

```python
# 建立其他职位表dataframe
df_market = pd.read_csv('市场.csv',engine='python')
df_technology = pd.read_csv('技术.csv',engine='python')
df_game = pd.read_csv('游戏.csv',engine='python')
df_function = pd.read_csv('职能.csv',engine='python')
df_create = pd.read_csv('设计.csv',engine='python')
df_run = pd.read_csv('运营.csv',engine='python')
df_sell = pd.read_csv('销售.csv',engine='python')
```

```python
# 多线程运行

import threading

thread1 = threading.Thread(target = add,args = (df_poduct,))
thread2 = threading.Thread(target = add,args =(df_market,))
thread3 = threading.Thread(target = add,args = (df_technology,))
thread4 = threading.Thread(target = add,args =(df_game,))
thread5 = threading.Thread(target = add,args = (df_function,))
thread6 = threading.Thread(target = add,args =(df_create,))
thread7 = threading.Thread(target = add,args = (df_run,))
thread8 = threading.Thread(target = add,args =(df_sell,))

thread1.start()
thread2.start()
thread3.start()
thread4.start()
thread5.start()
thread6.start()
thread7.start()
thread8.start()
thread1.join()
thread2.join()
thread3.join()
thread4.join()
thread5.join()
thread6.join()
thread7.join()
thread8.join()
# thread1.start()　忘删
#　thread1.join()
print('成功！')
```
